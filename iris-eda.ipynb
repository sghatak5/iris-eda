##### Understand the distribution of the data using Boxplot 
A box plot are used in statistics to graphically display various parameters at a glance. In a boxplot, the median , the interquartile range and the outliers can be read. A boxplot is often created to compare and contrast two or more groups. 

+ Box indicate the range in which the middle 50% of all the data lies
+ Lower end of the box in 1st quartile(q1) and the upper end is 3rd quartile(q3)
+ Between q1 and q3 is the interquartile range
+ Solid line indicates the mean
+ Black Line indicates the last point with outliers, which is 1.5 times the interquartile range. If there is no outlier, the black line is the max or min point. 
+ Circle is outliers

```python
df.boxplot()
```
